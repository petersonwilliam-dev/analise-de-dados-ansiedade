<h2><b>Fatores que influenciam aparição de ansiedade e depressão em indivíduos</b></h2>
<p>Essa análise busca entender como alguns fatores podem aumentar as chances de aparição de doenças mentais como depressão e ansiedade nas pessoas</p>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dados = pd.read_csv("./dados/anxiety_depression_data.csv")
dados
```

<h3><b>Pessoas que fazem terapia tem menos depressão e ansiedade? </b></h3>

<p>Vamos analisar!</p>

```python
terapia_gp = dados.groupby(['Therapy']).size().reset_index(name="quantidade de pessoas")
plt.figure(figsize=(8,6))
sns.barplot(x="Therapy", y="quantidade de pessoas", data=terapia_gp, palette='viridis')
plt.xlabel("Fazem terapia?")
plt.ylabel("quantidade de pessoas")
plt.title("Terapia ajuda na ausência de doenças mentais")
plt.xticks(ticks=[0, 1], labels=['Não', 'Sim'])
plt.show()
```

<h3><b>Meditação ajuda contra doenças mentais?</b></h3>

```python
meditation_gp = dados.groupby(['Meditation']).size().reset_index(name="Quantidade de pessoas")
plt.figure(figsize=(8,6))
sns.barplot(x="Meditation", y="Quantidade de pessoas", data=meditation_gp, palette='viridis')
plt.xlabel("Fazem meditação?")
plt.ylabel("Quantidade de pessoas")
plt.title("Meditação no combate a doenças mentais")
plt.xticks(ticks=[0,1], labels=['Não', 'Sim'])
plt.show()
```

```python
gender_gp = dados.groupby(['Gender']).size().reset_index(name="Quantidade")
plt.figure(figsize=(8,6))
sns.barplot(x="Gender", y="Quantidade", data=gender_gp, palette='viridis')
plt.xlabel("Gênero")
plt.ylabel("Quantidade de pessoas")
plt.title("Aparição de depressão e ansiedade de acordo com gêneros")
plt.show()
```

```python
gender_gp = dados.groupby(['Family_History_Mental_Illness']).size().reset_index(name="Quantidade")
plt.figure(figsize=(8,6))
sns.barplot(x="Family_History_Mental_Illness", y="Quantidade", data=gender_gp, palette='viridis')
plt.xlabel("Possui histórico de doenças mentais na família")
plt.ylabel("Quantidade de pessoas")
plt.title("Histórico familiar de doenças mentais contribuem")
plt.xticks(ticks=[0,1], labels=['Não', 'Sim'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x="Stress_Level", y="Depression_Score", data=dados)
plt.xlabel("Nível de estresse")
plt.ylabel("Nível de depressão")
plt.title("Estresse influencia na aparição de depressão")
plt.ticklabel_format(style="plain", axis='x')
plt.ticklabel_format(style="plain", axis='y')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x="Sleep_Hours", y="Depression_Score", data=dados)
plt.xlabel("Nível de estresse")
plt.ylabel("Nível de depressão")
plt.title("Estresse influencia na aparição de depressão")
plt.ticklabel_format(style="plain", axis='x')
plt.ticklabel_format(style="plain", axis='y')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x="Life_Satisfaction_Score", y="Depression_Score", data=dados)
plt.xlabel("Nível de satisfação na vida")
plt.ylabel("Nível de depressão")
plt.title("Estresse influencia na aparição de depressão")
plt.ticklabel_format(style="plain", axis='x')
plt.ticklabel_format(style="plain", axis='y')
plt.show()
```
